```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(df['Aggregate rating'], kde=True, bins=20)
plt.title('Distribution of Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
bins = [0, 1, 2, 3, 4, 5]
labels = ['1', '2', '3', '4', '5']
df['Rating category'] = pd.cut(df['Aggregate rating'], bins=bins, labels=labels, right=False)
rating_counts = df['Rating category'].value_counts()
print(rating_counts)
sns.barplot(x=rating_counts.index, y=rating_counts.values)
plt.title('Class Distribution of Aggregate Rating Categories')
plt.xlabel('Rating Category')
plt.ylabel('Count')
plt.show()
```
